```python
#Importing the required libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
#Read CSV
df = pd.read_csv("/content/drive/MyDrive/DV/NATURAL DISASTERS.csv")
print(df.info())
```

```python
#To Remove Unnamed Columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
print(df.info())
```

```python
#Replace Null Values with 0
df.fillna(value = 0,inplace = True)
df.isnull().sum()
df.head()
```

##HEAT MAP

```python

plt.title('null, missing and \'0\' values heatmap')
sns.heatmap((df.isnull()) | (df == 0), cmap = 'mako')
plt.show()
print('\nDUPLICATE VALUE COUNT: ', df.duplicated().sum())
```

##SWARM PLOT

```python
# groupby()- To group df based on Year
y=df.groupby('Year')
y
# Iterate through he Year column and their corresponding df
for Year,Year_df in y:
  print(Year) # Prints Key Value - Year
  print(Year_df) # Prints df corresponding to the key as value
```

##WORD CLOUD

```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/DV/NATURAL DISASTERS.csv")
comment_words = " "
stopwords = set(STOPWORDS)
#Iterating disaster subtype values
for val in df['Disaster Subtype'][0:3000]:
  val = str(val)
  tokens = val.split()
  #converting to lowercase
  for i in range(len(tokens)):
    tokens[i] = tokens[i].lower()
    comment_words += " ".join(tokens)+" "
#generate wordcloud
wc = WordCloud(width = 500, height = 500, background_color ='black', stopwords = stopwords, min_font_size = 10).generate(comment_words)
plt.figure(figsize = (8, 8), facecolor = None)

#display wordcloud
plt.imshow(wc)
plt.axis('off')
plt.tight_layout(pad = 0)
plt.show()
```

##JOINT PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/DV/NATURAL DISASTERS.csv")
#renaming column names
data=data.replace(['Hydrological'],'Hydro')
data=data.replace(['Biological'],'Bio')
data=data.replace(['Meteorological'],'Meteoro')
data=data.replace(['Climatological'],'Climat')
data=data.replace(['Geophysical'],'Geo')
data=data.replace(['Extra-terrestrial'],'Extra')
#plotting jointplot
sns.jointplot(x = 'Disaster Subgroup',y = 'Disaster Type',color='red',data = data)
```

##LINE PLOT

```python
#setting up size of the figure
plt.figure(figsize=[10,10])
state=df['Country'].iloc[1000:9000:1000] #selecting data to make dataframe state
ar=df['No Injured'].iloc[1000:9000:1000] #selecting data to make dataframe ar
disr=df['No Homeless'].iloc[1000:9000:1000] #selecting data to make dataframe disr
dr=df['Total Damages'].iloc[1000:9000:1000] #selecting data to make dataframe dr
plt.plot(state,ar,linestyle = "--",label='Number Injured') #plotting for the data in dataframe ar
plt.plot(state,disr,label='Number Homeless') #plotting for the data in dataframe disr
plt.plot(state,dr,label='Number Damages') #plotting for the data in dataframe dr
plt.plot(state,dr,linestyle = "-.",label='Total damage',color='red')
plt.xlabel('Disaster')
plt.ylabel('Affected')
plt.legend()
plt.show()
```
